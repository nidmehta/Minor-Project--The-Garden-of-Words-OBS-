```python
#Importing OpenCV python library
import cv2 #pip install opencv-contrib-python in Adminstrator: Anaconda prompt
```

```python
img = cv2.imread('C:/Users/nidhi/Book Recommendation System/FER 13/train/happy/IMG_20210823_193333_265.webp')
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img) #default colours are BGR
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB)) #To change to normal color
```

```python
from deepface import DeepFace #pip install deepface
```

```python
predictions = DeepFace.analyze(img, enforce_detection=False)
```

```python
predictions
```

```python
#type(predictions)
```

```python
predictions['dominant_emotion']
```

```python
faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml') #for the rectangle of position of face in the whole picture
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
#print(faceCascade.empty())
faces = faceCascade.detectMultiScale(gray, 1.1, 4)

#To draw a rectangle around the faces
for (x, y, w, h) in faces:
    cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 1)
```

```python
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
#Used for inserting text in a video
cv2.putText(
            img,
            predictions['dominant_emotion'],
            (0, 10),
            font, 
            0.5,
            (0, 0, 255),
            1,
            cv2.LINE_4
);
```
